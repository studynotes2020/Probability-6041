## Bernoulli process

Bernoulli process is a a sequence of independent Bernoulli trials. It is memoryless and discrete time.

When we study a distribution, we study its Random Variable, Expectation, Variance

If X is a bernoulli random variable

$X = \begin{cases} 
p \quad & \text{event happen}\\
1 - p \quad & \text{event not happen}
\end{cases}$

$$\begin{align}
E[X] & = \sum_x xp_X(x) \\
& = 1.p + 0.(1-p) \\
E[X] & = p
\end{align}$$

To compute variance, we use the formula

$$Var[X] = E[X^2] - (E[X])^2 = p - p^2$$

Because $X$ can only takes value of 0 and 1, $X^2$ is also 0 and 1 which implies $E[X^2] = E[X] = p$

What is the sample space for Bernoulli process?

Start with this question: what is this probability: $P(X_t = 1 \text{ for all } t)$, 0 < p < 1

$P(X_t = 1, \forall t) \leq P(X_1 = ... = X_k = 1) = p^k \leq \lim\limits_{k \to \infty} p^k = 0$

This is similar with continuous random variable where probability of each point is 0

So, the sample space of Bernoulli process is like the expansion of a real number. When you pick a real number in random, any particular real number has 0 probability 


### Number of successes S in n time slots

- Fixed the time, ask how many arrival

This is just binomial distribution

$P(S = k) = \binom{n}{k} p^k (1-p)^{n-k}$, k = 0, ..., n

$E[S] = np$

$Var(S) = np(1-p)$


### Interarrival times

- Fixed arrival, ask how many time does it take

This is geometric distribution

$T_1:$ number of trials until first success

$P(T_1 = t) = (1-p)^{t-1} p, \quad t = 1,2,...$

$E[T_1] = \frac{1}{p}$ which means if p is small, we take long time to success

$Var(T_1) = \frac{1-p}{p^2}$, derived in the textbook

If you buy a lottery ticket every day, what is the distribution of the length L of the first string of losing days? Is L geometric distribution?

No. Let's reason it. The shortest length L is 1, so by definition L cannot be zero. It means L + 1 cannot be 1. But if L + 1 were geometric, it could be equal to 1, therefore L + 1 is not geometric

Because one has no information about when the first failure will occur. To say start watching just before first failures mean the person has future insight about what is coming next. This means it is not an independent Bernoulli trial

However, if L starts after receiving first failure, and continue until the first success, then L would be a geometric distribution with parameter p

So, we talked about first arrival, we then continue to second and third arrival

$Y_3 = T_1 + T_2 + T_3$

Based on the success of the first time, we then start watching for second and third arrival. We have no foresight about the future, so they are independent Bernoulli trials and hence geometric distribution parameter p

The conclusion is the time until third arrival is the sum of 3 independent geometric random variable with the same parameter. This is true generally for k-th time arrival is the sum of k independent random variable

$Y_k = T_1 + ... T_k$, $T_i ~$ geometric(p)

Let say we are interested in how long it takes for a 100 customers to arrive. Take a geometric convolve it with another geometric, take the result, convolve it to another geometric. Do this 99th times, we get distribution of $Y_k$

Or we can find its probability mass function

$p_{Y_k} (t) = P(Y_k = t) = P(k-1 \text{ arrival in time [1,t-1] and arrival at time t})$

Time slots are independent, so these two events are independent -> can multiply

$= \binom{t-1}{k-1}p^{k-1}(1-p)^{t-k} p$. This is true for t >= k

Expected value of Yk is just k times expected value of T, so 

$E[Y_k] = \frac{k}{p}$


### Splitting of a Bernoulli Process

Split Bernoulli process into two queue

p for probability arrival, q for probability send to the first queue

First queue has probability pq, second queue has probability p(1-q)


### Merging of a Bernoulli Process

Record if arrived or not

P(arrival) = 1 - P(non-arrival) = $1 - (1-p)(1-q) = p + q - pq$
