# Review

- Number of arrivals in n time slots: binomial pmf

- Interarrival times: geometric pmf

- Time to k arrivals: Pascal pmf

$Y_k = T_1 + T_2 + ... + T_k$ where each T is independent geometric distribution with parameter p

$P(Y_k = t) = \binom{t-1}{k-1} p^k (1-p)^{t-k}, \quad t = k, k+1, ...$ which is called Pascal pmf



## Poisson Process

Poisson process is the continuous-time analog of Bernoulli process

$P(k, \tau) = $ probability of k arrivals in interval of duration $\tau$ (fixed) and assume that this probability is the same for all intervals of the same length $\tau$

$\sum_k P(k, \tau) = 1$ 

Numbers of arrivals in disjoint time intervals are independent

$P(k, \delta) = \begin{cases}
1 - \lambda\delta, & k = 0 \\
\lambda\delta, & k = 1 \\
0, & k > 1
\end{cases}$

$O(\delta^2)$ term is ignored

$\lambda$ is arrival rate and equals to expected number or arrivals per unit time

Another view of arrival rate is $$\lim\limits_{\delta \to 0}\frac{P(k, \tau)}{\delta} = \lambda$$

Expectations, E[number of arrival in [0, delta]] = $\lambda\delta . 1$

Now let take duration tau, divided by delta to get n intervals

$n = \frac{\tau}{\delta}$, at each interval we have probability of arrival $p = \lambda\delta \implies p = \frac{\lambda\tau}{n}$

From binomial pmf

$$\begin{align}
P(\text{k arrivals}) & = \binom{n}{k} p^k (1-p)^{n-k} \\
& = \binom{n}{k}(\frac{\lambda\tau}{n})^k(1- \frac{\lambda\tau}{n})^{n-k}\\
& = \frac{n!}{(n-k)!k!}.\frac{(\lambda\tau)^k}{n^k}.(1- \frac{\lambda\tau}{n})^{n}.(1- \frac{\lambda\tau}{n})^{-k} \\
\end{align} $$

Take the limit $n \to \infty$

$$(1- \frac{\lambda\tau}{n})^{n} \to e^{-\lambda\tau}, \quad (1- \frac{\lambda\tau}{n})^{-k} \to 1$$

Also, after n! has taken out (n-k)! terms, it left k terms, and

$$P(\text{k arrivals}) = \frac{n}{n}.\frac{n-1}{n}...\frac{n-k+1}{n}. \frac{(\lambda\tau)^k}{k!} e^{-\lambda\tau}$$

$$\frac{n}{n}.\frac{n-1}{n}...\frac{n-k+1}{n} \to 1$$

$$P(k, \tau) = \frac{(\lambda\tau)^k e^{-\lambda\tau}}{k!}, \quad k = 0, 1, ...$$

$E[N_t] = np = \lambda t$

$Var(N_t) = np(1-p) \to np = \lambda t$ when $\delta \to 0, p \to 0$

### Example

You get email according to a Poisson process at a rate of lambda = 5 messages per hour. You check your email every thirty minutes

- P(no new messages) = $P(0, 1/2) = \frac{(\lambda\tau)^0 e^{-\lambda\tau}}{0!}$

$\lambda t = 5 . \frac12 = 2.5$

$P(0, 1/2) = e^{-2.5}$

- P(one new message) = $P(1, 1/2) = \frac{(\lambda\tau)^1 e^{-\lambda\tau}}{1!} = 2.5e^{-2.5}$

Now let remind the usual interpretation of pdf times delta means success falls between $[t, t+\delta]$

$$f_{Y_k}(t)\delta = P(t \leq Y_k \leq t+\delta) = P(\text{k-1 arrivals in [0,t]}). \lambda \delta$$

$$f_{Y_k}(t) = \frac{(\lambda t)^{k-1} e^{-\lambda t}}{(k-1)!}.\lambda = \frac{\lambda^k t^{k-1} e^{-\lambda t}}{(k-1)!}$$

which is called Erlang distribution. It gives us the formula for the PDF of the time until k-th arrival. From the distribution we can see, more time is needed for more k arrivals

For k = 1 arrival

$$f_{Y_1}(t) = \lambda e^{-\lambda t}$$

and the expectation is geometric random variable with parameter lambda

$$E[Y_1] = \frac{1}{\lambda}$$

if one get average 2 call per hour, then one can expect wait half an hour for a call


### Merging Poisson Processes

Red bulb flashes (poisson process, $\lambda_1$) and Green bulb flashes (poisson process $\lambda_2$) merge into all flashes

What is the probability that the next arrival comes from the first process

We can divide into four scenarios and think of delta goes to zero

1. Red and Green

$\lambda_1\delta\lambda_2\delta \to 0$

2. Red and not Green

$\lambda_1\delta(1-\lambda_2\delta) \to \lambda_1\delta$

3. Not red and Green

$(1-\lambda_1\delta)\lambda_2\delta \to \lambda_2\delta$

4. Not red and not Green

$(1-\lambda_1\delta)(1-\lambda_2\delta) \to 1 - \lambda_1\delta-\lambda_2\delta$

So the probability of it flashes is $\lambda_1\delta + \lambda_2\delta$ and probability it is red is 

$$\frac{\lambda_1\delta}{\lambda_1\delta + \lambda_2\delta} = \frac{\lambda_1}{\lambda_1 + \lambda_2}$$
